```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
path = 'Data_Arthur/'
folder = 'paard_verticaal'
```

```python
mode  = folder.split('_')[0]

df_acc = pd.read_csv(f'{path}{folder}/Accelerometer.csv')
df_gyro = pd.read_csv(f'{path}{folder}/Gyroscope.csv')

if path == 'Data_Arthur/':
    df_lin_acc = pd.read_csv(f'{path}{folder}/Linear Accelerometer.csv')
else:
    df_lin_acc = pd.read_csv(f'{path}{folder}/Linear Acceleration.csv')

acc_columns = df_acc.columns
gyro_columns = df_gyro.columns
linacc_columns = df_lin_acc.columns
```

```python
# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(12, 12), sharex=True)

# Plotting Acceleration data
axs[0].plot(df_acc['Time (s)'], df_acc[acc_columns[1]], label='x', color='r')
axs[0].plot(df_acc['Time (s)'], df_acc[acc_columns[2]], label='y', color='g')
axs[0].plot(df_acc['Time (s)'], df_acc[acc_columns[3]], label='z', color='b')
axs[0].set_title(f'{mode} Acceleration')
axs[0].legend()
axs[0].set_ylabel('Acceleration (m/s^2)')

# Plotting Gyroscope data
axs[1].plot(df_gyro['Time (s)'], df_gyro[gyro_columns[1]], label='x', color='r')
axs[1].plot(df_gyro['Time (s)'], df_gyro[gyro_columns[2]], label='y', color='g')
axs[1].plot(df_gyro['Time (s)'], df_gyro[gyro_columns[3]], label='z', color='b')
axs[1].set_title(f'{mode} Gyroscope')
axs[1].legend()
axs[1].set_ylabel('Angular Velocity (rad/s)')

# Plotting Linear Acceleration data
axs[2].plot(df_lin_acc['Time (s)'], df_lin_acc[linacc_columns[1]], label='x', color='r')
axs[2].plot(df_lin_acc['Time (s)'], df_lin_acc[linacc_columns[2]], label='y', color='g')
axs[2].plot(df_lin_acc['Time (s)'], df_lin_acc[linacc_columns[3]], label='z', color='b')
axs[2].set_title(f'{mode} Linear Acceleration')
axs[2].legend()
axs[2].set_ylabel('Linear Acceleration (m/s^2)')
axs[2].set_xlabel('Time (s)')

# Adjust layout
plt.tight_layout()
plt.savefig(f'Plots/{mode}1.png')
plt.show()
```

```python
print("Missing values in Acceleration data:\n", df_acc.isnull().sum())
print("Missing values in Gyroscope data:\n", df_gyro.isnull().sum())
print("Missing values in Linear Acceleration data:\n", df_lin_acc.isnull().sum())
```

```python
# Plot heatmaps
fig, axs = plt.subplots(1, 3, figsize=(16, 4))

sns.heatmap(df_acc[[acc_columns[1], acc_columns[2], acc_columns[3]]].corr(),annot=True,cmap='coolwarm', ax=axs[0], vmin=-1, vmax=1)
axs[0].set_title(f'{mode} Acceleration Correlation Matrix')
axs[0].set_xticklabels([])

sns.heatmap(df_gyro[[gyro_columns[1], gyro_columns[2], gyro_columns[3]]].corr(),annot=True,cmap='coolwarm', ax=axs[1], vmin=-1, vmax=1)
axs[1].set_title(f'{mode} Gyroscope Correlation Matrix')
axs[1].set_xticklabels([])

sns.heatmap(df_lin_acc[[linacc_columns[1], linacc_columns[2], linacc_columns[3]]].corr(),annot=True,cmap='coolwarm', ax=axs[2], vmin=-1, vmax=1)
axs[2].set_title(f'{mode} Linear Acceleration Correlation Matrix')
axs[2].set_xticklabels([])

plt.tight_layout()
plt.savefig(f'Plots/heatmaps_{mode}1.png')
plt.show()
```

```python
sns.set(style="whitegrid")
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

sns.histplot(df_acc[acc_columns[1]], kde=True, ax=axs[0, 0], color='r')
sns.histplot(df_acc[acc_columns[2]], kde=True, ax=axs[0, 1], color='g')
sns.histplot(df_acc[acc_columns[3]], kde=True, ax=axs[0, 2], color='b')
axs[0, 0].set_title(f'{mode} Acceleration X Distribution')
axs[0, 1].set_title(f'{mode} Acceleration Y Distribution')
axs[0, 2].set_title(f'{mode} Acceleration Z Distribution')

sns.histplot(df_gyro[gyro_columns[1]], kde=True, ax=axs[1, 0], color='r')
sns.histplot(df_gyro[gyro_columns[2]], kde=True, ax=axs[1, 1], color='g')
sns.histplot(df_gyro[gyro_columns[3]], kde=True, ax=axs[1, 2], color='b')
axs[1, 0].set_title(f'{mode} Gyroscope X Distribution')
axs[1, 1].set_title(f'{mode} Gyroscope Y Distribution')
axs[1, 2].set_title(f'{mode} Gyroscope Z Distribution')

sns.histplot(df_lin_acc[linacc_columns[1]], kde=True, ax=axs[2, 0], color='r')
sns.histplot(df_lin_acc[linacc_columns[2]], kde=True, ax=axs[2, 1], color='g')
sns.histplot(df_lin_acc[linacc_columns[3]], kde=True, ax=axs[2, 2], color='b')
axs[2, 0].set_title(f'{mode} Linear Acceleration X Distribution')
axs[2, 1].set_title(f'{mode} Linear Acceleration Y Distribution')
axs[2, 2].set_title(f'{mode} Linear Acceleration Z Distribution')

plt.tight_layout()
plt.savefig(f'Plots/distributions_{mode}1.png')
plt.show()
```

```python
sns.pairplot(df_acc[[acc_columns[1], acc_columns[2], acc_columns[3]]])
plt.title(f'{mode} Acceleration Pairplot')
plt.savefig(f'Plots/acc_pairplots_{mode}1.png')
plt.show()

sns.pairplot(df_gyro[[gyro_columns[1], gyro_columns[2], gyro_columns[3]]])
plt.title(f'{mode} Gyroscope Pairplot')
plt.savefig(f'Plots/gyro_pairplots_{mode}1.png')
plt.show()

sns.pairplot(df_lin_acc[[linacc_columns[1], linacc_columns[2], linacc_columns[3]]])
plt.suptitle(f'{mode} Linear Acceleration Pairplot',y=1.02)
plt.savefig(f'Plots/linacc_pairplots_{mode}1.png')
plt.show()
```

```python
# Plot boxplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Acceleration
sns.boxplot(data=df_acc[[acc_columns[1],acc_columns[2],acc_columns[3]]], ax=axs[0])
axs[0].set_title(f'{mode} Acceleration')
# Gyroscope
sns.boxplot(data=df_gyro[[gyro_columns[1],gyro_columns[2],gyro_columns[3]]], ax=axs[1])
axs[1].set_title(f'{mode} Gyroscope')

# Linear Acceleration
sns.boxplot(data=df_lin_acc[[linacc_columns[1],linacc_columns[2],linacc_columns[3]]], ax=axs[2])
axs[2].set_title(f'{mode} Linear Acceleration')

plt.tight_layout()
plt.savefig(f'Plots/boxplots_{mode}1.png')
plt.show()
```
