```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
path = 'C:/Users/Nando Beijaard/Documents/GitHub/MLQS_g67/Own Code/'
folder = 'Data_Nando/metro_horizontaal_0506-1236'
```

```python
mode  = path.split('/')[-1].split('_')[0]

df_acc = pd.read_csv(f'{path}{folder}/Accelerometer.csv')
df_gyro = pd.read_csv(f'{path}{folder}/Gyroscope.csv')
df_lin_acc = pd.read_csv(f'{path}{folder}/Linear Acceleration.csv')
```

```python
# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(12, 12), sharex=True)

# Plotting Acceleration data
axs[0].plot(df_acc['Time (s)'], df_acc['Acceleration x (m/s^2)'], label='x', color='r')
axs[0].plot(df_acc['Time (s)'], df_acc['Acceleration y (m/s^2)'], label='y', color='g')
axs[0].plot(df_acc['Time (s)'], df_acc['Acceleration z (m/s^2)'], label='z', color='b')
axs[0].set_title('Acceleration')
axs[0].legend()
axs[0].set_ylabel('Acceleration (m/s^2)')

# Plotting Gyroscope data
axs[1].plot(df_gyro['Time (s)'], df_gyro['Gyroscope x (rad/s)'], label='x', color='r')
axs[1].plot(df_gyro['Time (s)'], df_gyro['Gyroscope y (rad/s)'], label='y', color='g')
axs[1].plot(df_gyro['Time (s)'], df_gyro['Gyroscope z (rad/s)'], label='z', color='b')
axs[1].set_title('Gyroscope')
axs[1].legend()
axs[1].set_ylabel('Angular Velocity (rad/s)')

# Plotting Linear Acceleration data
axs[2].plot(df_lin_acc['Time (s)'], df_lin_acc['Linear Acceleration x (m/s^2)'], label='x', color='r')
axs[2].plot(df_lin_acc['Time (s)'], df_lin_acc['Linear Acceleration y (m/s^2)'], label='y', color='g')
axs[2].plot(df_lin_acc['Time (s)'], df_lin_acc['Linear Acceleration z (m/s^2)'], label='z', color='b')
axs[2].set_title('Linear Acceleration')
axs[2].legend()
axs[2].set_ylabel('Linear Acceleration (m/s^2)')
axs[2].set_xlabel('Time (s)')

# Adjust layout
plt.tight_layout()
plt.savefig(f'{path}/Plots_Nando/{mode}1.png')
plt.show()
```

```python
print("Missing values in Acceleration data:\n", df_acc.isnull().sum())
print("Missing values in Gyroscope data:\n", df_gyro.isnull().sum())
print("Missing values in Linear Acceleration data:\n", df_lin_acc.isnull().sum())
```

```python
# Plot heatmaps
fig, axs = plt.subplots(1, 3, figsize=(16, 4))

sns.heatmap(df_acc[['Acceleration x (m/s^2)', 'Acceleration y (m/s^2)', 'Acceleration z (m/s^2)']].corr(),annot=True,cmap='coolwarm', ax=axs[0], vmin=-1, vmax=1)
axs[0].set_title('Acceleration Correlation Matrix')
axs[0].set_xticklabels([])

sns.heatmap(df_gyro[['Gyroscope x (rad/s)', 'Gyroscope y (rad/s)', 'Gyroscope z (rad/s)']].corr(),annot=True,cmap='coolwarm', ax=axs[1], vmin=-1, vmax=1)
axs[1].set_title('Gyroscope Correlation Matrix')
axs[1].set_xticklabels([])

sns.heatmap(df_lin_acc[['Linear Acceleration x (m/s^2)', 'Linear Acceleration y (m/s^2)', 'Linear Acceleration z (m/s^2)']].corr(),annot=True,cmap='coolwarm', ax=axs[2], vmin=-1, vmax=1)
axs[2].set_title('Linear Acceleration Correlation Matrix')
axs[2].set_xticklabels([])

plt.tight_layout()
plt.savefig(f'{path}/Plots_Nando/heatmaps_{mode}1.png')
plt.show()
```

```python
sns.set(style="whitegrid")
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

sns.histplot(df_acc['Acceleration x (m/s^2)'], kde=True, ax=axs[0, 0], color='r')
sns.histplot(df_acc['Acceleration y (m/s^2)'], kde=True, ax=axs[0, 1], color='g')
sns.histplot(df_acc['Acceleration z (m/s^2)'], kde=True, ax=axs[0, 2], color='b')
axs[0, 0].set_title('Acceleration X Distribution')
axs[0, 1].set_title('Acceleration Y Distribution')
axs[0, 2].set_title('Acceleration Z Distribution')

sns.histplot(df_gyro['Gyroscope x (rad/s)'], kde=True, ax=axs[1, 0], color='r')
sns.histplot(df_gyro['Gyroscope y (rad/s)'], kde=True, ax=axs[1, 1], color='g')
sns.histplot(df_gyro['Gyroscope z (rad/s)'], kde=True, ax=axs[1, 2], color='b')
axs[1, 0].set_title('Gyroscope X Distribution')
axs[1, 1].set_title('Gyroscope Y Distribution')
axs[1, 2].set_title('Gyroscope Z Distribution')

sns.histplot(df_lin_acc['Linear Acceleration x (m/s^2)'], kde=True, ax=axs[2, 0], color='r')
sns.histplot(df_lin_acc['Linear Acceleration y (m/s^2)'], kde=True, ax=axs[2, 1], color='g')
sns.histplot(df_lin_acc['Linear Acceleration z (m/s^2)'], kde=True, ax=axs[2, 2], color='b')
axs[2, 0].set_title('Linear Acceleration X Distribution')
axs[2, 1].set_title('Linear Acceleration Y Distribution')
axs[2, 2].set_title('Linear Acceleration Z Distribution')

plt.tight_layout()
plt.savefig(f'{path}/Plots_Nando/distributions_{mode}1.png')
plt.show()
```
